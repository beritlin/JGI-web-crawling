### crawling fungal images on JGI website

```python
driver_path = "/Users/pei-yu/Desktop/chromedriver"
driver = webdriver.Chrome(driver_path)
link="https://genome.jgi.doe.gov/portal/"
driver.get(link)
time.sleep(1)


for i in range(lenname[0]):
    try:
        sp=funname.iloc[i]
        inputsp = driver.find_element_by_id('query')
        inputsp.send_keys(sp[0])
        search = driver.find_element_by_id('submit')
        search.click() 
        time.sleep(1)

        genome = driver.find_element_by_xpath('/html/body/div[3]/div/div[5]/div/div[2]/div/div/table/tbody/tr[2]/td[3]/div[1]/div/a[2]')
        genome.click()
        time.sleep(1)
    
        html = driver.page_source 
        soup = BeautifulSoup(html,"html.parser")
        image = soup.find('img',{'class':'tmpl_img'})
    
        img = str('https://mycocosm.jgi.doe.gov/'+image.get('src'))
        x = img.split(';')
        url = x[0]
        print(i,url)
        shutil(url)

        
        home = driver.find_element_by_xpath('//*[@id="jgi-header"]/nav/ul/li[3]/a')
        home.click()
        time.sleep(1)
    except:
        print(i,'NA')
        link="https://genome.jgi.doe.gov/portal/"
        driver.get(link)
        time.sleep(1)
        pass
driver.close()
```
